```python
def sodium_cleaning(value):
    if value == '-':
        return None
    return float(value)

df['sodium (g)'] = df['sodium (g)'].apply(sodium_cleaning)
```

```python
def vitaminA_cleaning(value):
    if value == '-':
        return 0
    else:
        return float(value)

df['vitamin A (IU)'] = df['vitamin A (IU)'].apply(vitaminA_cleaning)
```

```python
df.info()
```

```python
def convert_b6(value):
    result = value.replace('-', '0.033')
    return float(result)

df['vitamin B6 (mg)'] = df['vitamin B6 (mg)'].apply(convert_b6)
df['vitamin B6 (mg)'].value_counts()
```

```python
def convert(value):
    if '-' in value:
        result = value.replace('-','0')
        return float(result)
    else:
        return float(value)
df['vitamin B2 (mg)'] = df['vitamin B2 (mg)'].apply(convert)
```

```python
def convert(value):
    if '-' in value:
        result = value.replace('-','0')
        return float(result)
    else:
        return float(value)
df['vitamin B1 (mg)'] = df['vitamin B1 (mg)'].apply(convert)
```

```python
def convert_magnessium(value):
    if value == '-':
        del(value)
    else:
        return float(value)

df['magnessium (mg)'] = df['magnessium (mg)'].apply(convert_magnessium)
df['potassium (mg)'] = df['potassium (mg)'].apply(convert_magnessium)
```

```python
def convertFloat(value):
    if value != "-":
        return float(value)
    else:
        return 0

df["vitamin E (mg)"] = df["vitamin E (mg)"].apply(convertFloat)
```
