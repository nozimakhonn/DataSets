```python
import pandas as pd
df = pd.read_csv('diabetes.csv')
```

```python
#Hypothesis
#1. People with low physical activity levels are more likely to develop insulin resistance. 
#2. A higher Body Mass Index (BMI) comes from unhealthy dietary habits that can cause diabetes. 
#3. Abnormal insulin levels raise the chance of weight Gain During Pregnancy can later develop into diabetes. 
#4. Higher blood Pressure levels are closely linked to environmental Factors. 
#5. A history of pregnancy, both complications and normal factors, increases the risk of high Cholesterol Levels. 
#6. Digestive Enzyme Levels can cause to rise in blood pressure more than usual. 
```

```python
df.info()
```

```python
#1
physical = df.groupby('Physical Activity')['Insulin Levels'].mean().sort_values(ascending=False)
df.plot(x = 'Physical Activity', y='Insulin Levels', kind = 'scatter')
```

```python
#2
def converting(diet):
    if "Healthy" in diet:
        return "Healthy"
    else:
        return "Unhealthy"
df["Category"] = df['Dietary Habits'].apply(converting)
diets = df.groupby("Category")['BMI'].mean()
diets.plot(kind='pie', grid=True, figsize=(8, 5))
```

```python
#3
preg = df.groupby('Insulin Levels')['Weight Gain During Pregnancy'].mean().sort_values(ascending=False)
df.plot(x = 'Insulin Levels', y='Weight Gain During Pregnancy', kind = 'scatter')
```

```python
#4
def convert(fact):
    if "Present" in fact:
        return "Present"
    else:
        return "Absent"
df["Factors"] = df['Environmental Factors'].apply(convert)
fbl = df.groupby("Factors")['Blood Pressure'].mean()
fbl.plot(kind='barh', grid=True, figsize=(8, 5))
```

```python
#5
def smth(blood):
    if "Normal" in blood:
        return "Normal"
    else:
        return "Complications"
df["History"] = df['Pregnancy History'].apply(smth)
bl = df.groupby("History")['Cholesterol Levels'].mean()
bl.plot(kind='bar', grid=True, figsize=(8, 5))
```

```python
#6
digest = df.groupby('Digestive Enzyme Levels')['Blood Pressure'].mean().sort_values(ascending=False)
df.plot(x = 'Digestive Enzyme Levels', y='Blood Pressure', kind = 'scatter')
```
