```python
def genetic_cleaning(value):
    try:
        value = str(value)
        if 'Positive' in value:
            return 1
        return int(value)
    except:
        return 0

df['Genetic Markers'] = df['Genetic Markers'].apply(genetic_cleaning)
def aut_cleaning(value):
    try:
        value = str(value)
        if 'Positive' in value:
            return 1
        return int(value)
    except:
        return 0

df['Autoantibodies'] = df['Autoantibodies'].apply(aut_cleaning)
def fam_cleaning(value):
    try:
        value = str(value)
        if 'No' in value:
            return 1
        return int(value)
    except:
        return 0

df['Family History'] = df['Family History'].apply(fam_cleaning)
def env_cleaning(value):
    try:
        value = str(value)
        if 'Absent' in value:
            return 1
        return int(value)
    except:
        return 0

df['Environmental Factors'] = df['Environmental Factors'].apply(env_cleaning)
def phys_cleaning(value):
    try:
        value = str(value)
        if 'Low' in value:
            return 1
        return int(value)
    except:
        return 0

df['Physical Activity'] = df['Physical Activity'].apply(phys_cleaning)
def diet_cleaning(value):
    try:
        value = str(value)
        if 'Unhealthy' in value:
            return 1
        return int(value)
    except:
        return 0

df['Dietary Habits'] = df['Dietary Habits'].apply(diet_cleaning)
def diet_cleaning(value):
    try:
        value = str(value)
        if 'Healthy' in value:
            return 1
        return int(value)
    except:
        return 0

df['Dietary Habits'] = df['Dietary Habits'].apply(diet_cleaning)
```

```python
def diet_cleaning(value):
    try:
        value = str(value)
        if 'High Risk' in value:
            return 0
        return int(value)
    except:
        return 1

df['Ethnicity'] = df['Ethnicity'].apply(diet_cleaning)
def sm_cleaning(value):
    try:
        value = str(value)
        if 'Smoker' in value:
            return 0
        return int(value)
    except:
        return 1

df['Smoking Status'] = df['Smoking Status'].apply(sm_cleaning)
def gl_cleaning(value):
    try:
        value = str(value)
        if 'Normal' in value:
            return 0
        return int(value)
    except:
        return 1

df['Glucose Tolerance Test'] = df['Glucose Tolerance Test'].apply(gl_cleaning)
def his_cleaning(value):
    try:
        value = str(value)
        if 'Yes' in value:
            return 0
        return int(value)
    except:
        return 1

df['History of PCOS'] = df['History of PCOS'].apply(his_cleaning)
def preg_cleaning(value):
    try:
        value = str(value)
        if 'Normal' in value:
            return 0
        return int(value)
    except:
        return 1

df['Pregnancy History'] = df['Pregnancy History'].apply(preg_cleaning)
def sys_cleaning(value):
    try:
        value = str(value)
        if 'Yes' in value:
            return 0
        return int(value)
    except:
        return 1

df['Cystic Fibrosis Diagnosis'] = df['Cystic Fibrosis Diagnosis'].apply(sys_cleaning)
def st_cleaning(value):
    try:
        value = str(value)
        if 'Yes' in value:
            return 0
        return int(value)
    except:
        return 1

df['Steroid Use History'] = df['Steroid Use History'].apply(st_cleaning)
def gen_cleaning(value):
    try:
        value = str(value)
        if 'Positive' in value:
            return 1
        return int(value)
    except:
        return 0

df['Genetic Testing'] = df['Genetic Testing'].apply(gen_cleaning)
def liv_cleaning(value):
    try:
        value = str(value)
        if 'Normal' in value:
            return 0
        return int(value)
    except:
        return 1

df['Liver Function Tests'] = df['Liver Function Tests'].apply(liv_cleaning)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
x = df.drop('Insulin Levels', axis=1)
y = df['Insulin Levels']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)#mashtab kamaytiradi
```

```python
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train, y_train)
```

```python
predictions = classifier.predict(x_test)
```

```python
print('Percentage %:', accuracy_score(y_test, predictions) * 100)
```

```python
print('Confusion matrix:\n', confusion_matrix(y_test, predictions))
```
